```python
    def plot_methods(to_plot, types="steps", ax=None):

        colors = ["green", "blue", "black", "red", "orange", "yellow"]
        #a_dict = {"lr":[],"steps":[],"time":[]}

        for i in range(0, len(to_plot)):

            res = to_plot[i]["data"][types].strip('][').split(', ')
            # Result and its type
            
            tmp = {"target":[], "x":[]}
            counter = 0
            for x in res:
                x = x.replace("[", "")
                x = x.replace("]", "")
    
                if counter % 2 == 0:
                    tmp["target"].append(round(float(x) / 20 * 100,2))
                else:
                    tmp["x"].append(float(x))
                counter +=1

            ax.scatter(    tmp["x"],
                            tmp["target"] ,
                            c=colors[i], 
                            label=to_plot[i]["alg"])

        
        ax.set_ylabel("RMSE voči škále hodnotenia v [%]")

        if types == "steps":
            ax.set_xlabel("Počet Krokov")
        elif types == "time":
            ax.set_xlabel("Čas v [s]")
        else :
            ax.set_xlabel("Parameter učenia sa")

        ax.legend()


    def call_plots_methods(to_plot=None, evaluations=False, lr=False, time=False, save_path=None):
        
        what_plot = []
        min = None
        if lr:
            what_plot.append({'type':"lr"})
        if evaluations:

            what_plot.append({'type':"steps"})
        if time:
            what_plot.append({'type':"time"})
            
        figure, axes = plt.subplots(
                            nrows=1, ncols=len(what_plot), constrained_layout=True,
                            figsize=(10, 5)
                            )  
          
        for i in range(0, len(what_plot)):
            plot_methods(to_plot=to_plot, types=what_plot[i]["type"], ax=axes[i])

        if save_path != None:
            plt.savefig(save_path)

    ## actualy not working but for future  using

    def graphs_alghoritms(df, all=False, sgd=False, svd=False, als=False,
                          values_sgd = None, values_als = None, values_svd = None,
                          eval=False, lr=False, time=False, save_path=None, print=False):

        to_plot = []
    

        
        to_plot.append({"alg":"als","data":df["als"]})

        to_plot.append({"alg":"svd","data":df["svd"]})
        to_plot.append({"alg":"sgd","data":df["sgd"]})

        call_plots_methods(to_plot=to_plot,evaluations=True, lr=True, time=True, save_path="jokes_hyper.png")
```
